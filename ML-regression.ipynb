```python
import pandas as pd
import numpy as np
import tensorflow as tf

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set(style="darkgrid")
```

```python
petrol_cons = pd.read_csv('petrol_consumption.csv')
petrol_cons.head()
```

```python
information_data = petrol_cons.iloc[:, 0:4].values
results = petrol_cons.iloc[:, 4].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# o train_test_split pega os dados e separa o que vai ser treino, o que vai ser test. Falando que 20% é teste
train_info, test_info, results_train, result_test = train_test_split(
  information_data, 
  results, 
  test_size=0.2, 
  random_state=0
)

# o StandardScaler é usado para deixar os valor com o mesmo peso ACHO
sc = StandardScaler()
train_info = sc.fit_transform(train_info)
test_info = sc.transform(test_info)
```

```python
                                                                                              import os
                                                                                              os.environ['CUDA_VISIBLE_DEVICES'] = "0"

from tensorflow.keras.layers import Input, Dense, Activation,Dropout
from tensorflow.keras.models import Model

input_layer = Input(shape=(information_data.shape[1],))
dense_layer_1 = Dense(100, activation='relu')(input_layer)
dense_layer_2 = Dense(50, activation='relu')(dense_layer_1)
dense_layer_3 = Dense(25, activation='relu')(dense_layer_2)
output = Dense(1)(dense_layer_3)

model = Model(inputs=input_layer, outputs=output)
model.compile(loss="mean_squared_error" , optimizer="adam", metrics=["mean_squared_error"])
```

```python
print(model.summary())
```

```python
history = model.fit(train_info, results_train, batch_size=2, epochs=100, verbose=1, validation_split=0.2)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

pred_train = model.predict(train_info)
print(np.sqrt(mean_squared_error(results_train,pred_train)))

pred = model.predict(test_info)
print(np.sqrt(mean_squared_error(result_test,pred)))
```

```python
score = model.evaluate(test_info, result_test, verbose=1)


print("Test Score:", score[0])
print("Test Accuracy:", score[1])
```
